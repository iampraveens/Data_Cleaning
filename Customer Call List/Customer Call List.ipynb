```python
import pandas as pd
```

```python
data = pd.read_excel('Customer Call List.xlsx')
# Read the data from local
data
```

```python
data.isnull().sum() 
# check for Null values
```

```python
data.dtypes 
# check for data types
```

```python
data = data.drop(columns='Not_Useful_Column') 
# Drop unwanted column
```

```python
data
# After dropping column 'Not_Useful_Column'
```

```python
data['Last_Name'] = data['Last_Name'].str.replace('/', '')
data['Last_Name'] = data['Last_Name'].str.replace('...', '')
data['Last_Name'] = data['Last_Name'].str.replace('_', '')
# Removing unwanted special characters for 'Last_Name' column
data
```

```python
data['Phone_Number'] = data['Phone_Number'].str.replace('/', '')
data['Phone_Number'] = data['Phone_Number'].str.replace('-', '')
data['Phone_Number'] = data['Phone_Number'].str.replace('|', '')
# Removing unwanted special characters for 'Phone_Number' column
data
```

```python
data['Phone_Number'] = data['Phone_Number'].apply(lambda x: str(x))
# converting 'Phone_Number' column into string to avoid error while formating the phone numbers
# ex. 1234567899 -> 123-456-7899
```

```python
data['Phone_Number'] = data['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# Formating the phone numbers
```

```python
data
```

```python
data['Phone_Number'] = data['Phone_Number'].str.replace('nan--', '')
data['Phone_Number'] = data['Phone_Number'].str.replace('Na--', '')
# Removing NaN values from 'Phone_Number' column
data
```

```python
data['Address'].str.split(',', expand=True)
# Splitting the 'Address' column by comma separated 
```

```python
data[['Street_Address', 'State', 'Zip_Code']] = data['Address'].str.split(',', expand=True)
# Create columns and Append splitted Address into actual Dataframe
data
```

```python
data = data.drop(columns='Address') 
# Dropping 'Address' column, as we don't need it anymore
data
```

```python
data['Paying Customer'] = data['Paying Customer'].str.replace('Yes', 'Y')
data['Paying Customer'] = data['Paying Customer'].str.replace('No', 'N')
# Formating Yes and No in 'Paying Customer'
data
```

```python
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('Yes', 'Y')
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('No', 'N')
data
```

```python
data = data.fillna('')
# Filling up NaN, None values as empty
data
```

```python
for drop_yes in data.index:
    if data.loc[drop_yes, 'Do_Not_Contact'] == 'Y':
        data.drop(drop_yes, inplace=True)
# Dropping Customers, who we not gonna call again
data
```

```python
for empty_numbers in data.index:
    if data.loc[empty_numbers, 'Phone_Number'] == '':
        data.drop(empty_numbers, inplace=True)
# Dropping Customers, who's phone numbers doesn't exists
data

# Another way to drop Customers, who's phone numbers doesn't exists
# data = data.dropna(subset='Phone_Number'), inplace=True)
```

```python
data = data.drop(columns= 'Zip_Code')
# Drop 'Zip_Code' column
data
```

```python
data = data.drop_duplicates() 
# Drop duplicates, forgot to do it at initial 
data
```

```python
data = data.reset_index(drop=True)
# Resetting the index values
data
```
